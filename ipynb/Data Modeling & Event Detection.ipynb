```python
import numpy as np
import scipy as sp
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from google.colab import drive

drive.mount('/content/drive')
data = pd.read_csv('/content/drive/MyDrive/CS 132/NewData.csv', nrows = 150)
NewData = data.dropna(axis = 1)
data
```

```python
column_names = ['Likes', 'Replies', 'Retweets', 'Quote Tweets']
NewData['Engagements'] = NewData[column_names].sum(axis=1)
NewData['Date posted'] = pd.to_datetime(NewData['Date posted'])
```

```python
TSA = pd.DataFrame({'Tweet': NewData['Tweet'], 'Date': NewData['Date posted'], 'Engagements': NewData['Engagements']})
TSA['Tweet'] = 1
TSA['Year'] = TSA['Date'].dt.year
TSA['Month'] = TSA['Date'].dt.month
TSA['Day'] = TSA['Date'].dt.day
TSA['Engagements'] = TSA.pop('Engagements')
TSA = TSA.drop(['Day'], axis='columns')
TSA
```

```python
nTSA = TSA.groupby(by=['Month', 'Year']).agg({'Tweet': ['sum'],'Engagements': ['sum'], 'Date': min})
nTSA = nTSA.reset_index()
nTSA.columns = ['_'.join(col) for col in nTSA.columns.values]
mapping = {nTSA.columns[0]: 'Month', nTSA.columns[1]: 'Year', nTSA.columns[2]: 'Tweet', nTSA.columns[3]: 'Engagements', nTSA.columns[4]: 'Date'}
nTSA = nTSA.rename(columns=mapping)

nTSA.sort_values(by='Date',inplace = True)
nTSA.reset_index(drop = True, inplace = True)
nTSA
```

# Linear Regression

```python
## Perform regression modeling
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import r2_score, mean_squared_error

import statsmodels.api as sm
TSA_LR = nTSA

# Convert datetime to int
x = TSA_LR['Date'].astype(int) / 10**9  # Convert to seconds (UNIX epoch start)
x = x.values.reshape(-1, 1)
y = TSA_LR['Tweet']

#--------------------------------------------------------------
# Linear regression model
#--------------------------------------------------------------
# Stastical approach
x_lms = sm.add_constant(x)
linear_model_stat = sm.OLS(y, x_lms)
lms_results = linear_model_stat.fit()
p_values = lms_results.pvalues[1:]

# Machine learning approach (no p-values)
linear_model = LinearRegression()
linear_model.fit(x_lms, y)
y_linear_pred = linear_model.predict(x_lms)

# Calculate R2 and RMSE for linear regression model
linear_r2 = r2_score(y, y_linear_pred)
linear_rmse = np.sqrt(mean_squared_error(y, y_linear_pred))

print("Model Evaluation")
print("\nLinear Regression: RMSE=%.2f, R2=%.2f" % (linear_rmse, linear_r2))
for i, p_value in enumerate(p_values.index):
  print(f'P({p_value}): {p_values[i]}')

if any(p_values <= 0.05):
  print("There is a significant relationship between the predictor and the response\n")
else:
  print("There is no significant relationship between the predictor and the response\n")


# Plot the model
import plotly.graph_objects as go

xtt = TSA_LR['Date'].dt.strftime('%Y-%m-%d')
scatter_actual = go.Scatter(x=xtt, y=y, mode='markers', name='Actual', marker=dict(color='blue', opacity=0.3))

line_regression = go.Scatter(x=xtt, y=y_linear_pred, mode='lines', name='LR', line=dict(color='red', dash='dash'))

data = [scatter_actual, line_regression]

layout = go.Layout(xaxis=dict(title=dict(text='Date',font=dict(size=30),),tickfont=dict(
            size=20,  # Set the desired font size for the ticks
        )),
                   yaxis=dict(title=dict(text='Number of Tweets',font=dict(size=30)),tickfont=dict(
            size=20,  # Set the desired font size for the ticks
        )),
                   title='Linear Regression Model for Twitter Marcos Chatter',
                   showlegend=True,
                   height=600)
fig = go.Figure(data=data, layout=layout)
fig.update_layout(
    title=dict(
        text='Linear Regression Model for Twitter Marcos Chatter',
        font=dict(
            size=30  # Set the desired font size for the title
        )
    )
)
fig.update_layout(legend=dict(title_font_family="Times New Roman",
                              font=dict(size= 20)
))
fig.show()
```

```python
## Perform peak detection
from scipy.signal import find_peaks

# Convert datetime to int
x = TSA_LR['Date'].astype(int) / 10**9  # Convert to seconds (UNIX epoch start)
x = x.values.reshape(-1, 1)
y = TSA_LR['Tweet']

peaks, _ = find_peaks(TSA_LR['Tweet'],
                      height=1,  # height of peaks
                      width=0.1,       # width of peaks
                      threshold=1, # vertical distance to its neighboring samples
                      distance=1,   # minimal horizontal distance (>= 1) in samples between neighbouring peaks
                      prominence=1) # vertical distance between the peak and its lowest contour line


# Extract event timestamps
events = TSA_LR.iloc[peaks]

# Plot peaks
fig = go.Figure()

# Original data
fig.add_trace(go.Scatter(
  x=TSA_LR['Date'].dt.strftime('%Y-%m-%d'),
  y=TSA_LR['Tweet'],
  mode='lines',
  name='Original Data'))

# Peaks
fig.add_trace(go.Scatter(
  x=events['Date'].dt.strftime('%Y-%m-%d'),
  y=events['Tweet'],
  mode='markers',
  name='Peaks',
  marker=dict(
    color='red',
    size=8,
    symbol='x')))

fig.update_layout(height=600,
                  title='Twitter Marcos Chatter Peak Events', title_font_size= 30,
                  xaxis = dict(title=dict(text='Date',font=dict(size=30)), tickfont=dict(
            size=20,  # Set the desired font size for the ticks
        )),
                  yaxis = dict(title=dict(text='Number of Tweets',font=dict(size=30)), tickfont=dict(
            size=20,  # Set the desired font size for the ticks
        )))
xtv = np.arange(0,16,1)
xtt = TSA_LR['Date'].dt.strftime('%Y-%m-%d')[xtv]
fig.update_xaxes(tickmode='array',
                 ticktext=xtt)

fig.show()
```
